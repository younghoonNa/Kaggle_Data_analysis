```r
install.packages("extrafont") 
library(extrafont)
system("sudo apt-get install fonts-nanum-coding") 
font_import(pattern = "Nanum")
```

```r
par(oma = c(0,0,2,0))
x <- c(1,3,5,2,4)
barplot(x, xlab = "x label", ylab = "y label")
axis(side = 1, at = 1:5, label = LETTERS[1:5], line = 2)
axis(side = 2)
axis(side = 3)
axis(side = 4) 
```

```r
x <- c(1,3,5,2,4)
barplot(x, xlab = "x label", ylab = "y label")
title(main = "Main title 1", line = 1)
title(main = "Main title 2", line = 2)
title(main = "Main title 3", line = 3)
```

```r
options(repr.plot.width=13, repr.plot.height=10)

par(mar = c(4,4,4,4), oma = c(4,0,0,0))
x <- c(1,3,5,2,4)
barplot(x, xlab = "", ylab = "")
title (xlab = "x label 1", line = 1)
title (xlab = "x label 2", line = 2)
title (xlab = "x label 3", line = 3)

title (ylab = "y label 1", line = 1)
title (ylab = "y label 2", line = 2)
title (ylab = "y label 3", line = 3)

title (sub = "sub Titel 1", line = 1, outer = T)
title (sub = "sub Title 2", line = 2, outer = T)
title (sub = "sub Title 3", line = 3, outer = T)

```

```r
par(mar = c(0,0,2,0))
plot(1:20, 1:20, type = 'n', xlab = "x lab", ylab = "ylab", main = "Text")

points(5, 15, pch = 16)
text(5,15, adj = 0, label = "(5,15) Computer")

points(5, 10, pch = 16)
text(5, 10, adj = 0, label = "adj = 0 Computer")

points(5, 12, pch = 16)
text(5, 12, label = "Basic")

points(5, 8, pch = 16)
text(5,8, adj = 0.3, label = "(5,8) Computer")

points(5, 6, pch = 16)
text(5,6, adj = 0.7, label = "(5,6) Computer")

points(5, 4, pch = 16)
text(5, 4, adj = 1, label = "adj = 1 Computer")

#pos 적용
points(15, 15, pch = 16)
text(15,15, pos=1, "pos = 1 : 아래쪽")

points(15, 15, pch = 16)
text(15,15, pos=2, "pos = 2 : 아래쪽")

points(15, 15, pch = 16)
text(15,15, pos=3, "pos = 3 : 아래쪽")

points(15, 15, pch = 16)
text(15,15, pos=4, "pos = 4 : 아래쪽")

#offset
points(15,5, pch=16)
text(15,5, pos = 1, offset = 1, "pos = 1 : 이래쪽, offset = 1")
text(15,5, pos = 2, offset = 2, "pos = 2 : 왼쪽, offset = 2")
text(15,5, pos = 3, offset = 3, "pos = 3 : 위쪽, offset = 3")
text(15,5, pos = 4, offset = 4, "pos = 4 : 오른쪽, offset = 4")
```

```r
par(mar = c(4,4,4,4), oma = c(4,0,0,0))
x <- c(1,3,5,2,4)
barplot(x, xlab = "x label", ylab = "y label")
mtext("Outer side 1", side = 1, line = 1, outer = T)
mtext("Outer side 2", side = 1, line = 2, outer = T)
mtext("Outer side 2", side = 1, line = 3, outer = T)

mtext("Side 1 line = 1, adj = 0 : 왼쪽", side = 1, line = 1, adj = 0)
mtext("Side 1 line = 2, adj = 0.5 : 중앙", side = 1, line = 2, adj = 0.5)
mtext("Side 1 line = 3, adj = 1 : 오른쪽", side = 1, line = 3, adj = 1)

mtext("Side 2 line = 1, adj = 0 : 왼쪽", side = 2, line = 1, adj = 0)
mtext("Side 2 line = 2, adj = 0.5 : 중앙", side = 2, line = 2, adj = 0.5)
mtext("Side 2 line = 3, adj = 1 : 오른쪽", side = 2, line = 3, adj = 1)
```

```r
plot(1:10, 1:10, type = 'n', xlab = "x label", ylab = "y label", main = "Legend")
twodata <- c("Data1", "Data2")
legend(2,8, legend = twodata, pch = 1:2, col = 1:2, title = "점")
legend(4,8, legend = twodata, lty = 1:2, col = 1:2, title = "선")
legend(7,8, legend = twodata, pch = 1:2, lty = 1:2, col = 1:2, title = "점과 선")
legend(2,4, legend = twodata, fill = 1:2, title = "면")
legend(6,4, legend = twodata, fill = 1:2, density = 30, title = "면과 밀도")
```

```r
plot(1:10, 1:10, type = "n", xlab = "xlab", ylab = "y lab", main = "Legend")
twodata <- c("Data1", "Data2")
legend("topleft", legend = twodata, pch = 1:2, col = 1:2, title = "점")
legend("center", legend = twodata, lty = 1:2, col = 1:2, title = "선")
legend("topright", legend = twodata, pch = 1:2, lty = 1:2, col = 1:2, title = "점과 선")
legend("bottomleft", legend = twodata, fill = 1:2, title = "면")
legend("bottomright", legend = twodata, fill = 1:2, density = 30, title = "면과 밀도")
```
