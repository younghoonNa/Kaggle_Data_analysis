```r
par(mfrow = c(1,3))
vec01 <- c(1,3,5,2,4)
name <- c(1,2,3,4,5)
barplot(vec01, names = name, main = "density = NULL", density = NULL)
barplot(vec01, names.arg = name , main = "density = 5", density = 5)
barplot(vec01, names = name , main = "density = 10", density = 10)
```

```r
par(mfrow = c(1,3))
vec01 <- c(1,3,5,2,4)
barplot(vec01, names = c(1,2,3,4,5), density = 10, angle = 30, main = "angle = 30")
barplot(vec01, names.arg = c(1,2,3,4,5), density = 10, angle =60, main = "angle = 60")
barplot(vec01, names.arg = c(1,2,3,4,5), density = 10, angle = 135, main = "angle = 135")
```

```r
par(mfrow = c(1,3))
vec01 <- c(1,3,5,2,4)
barplot(vec01, names = c(1,2,3,4,5))
title("default color")
barplot(vec01, names = c(1,2,3,4,5), col = rainbow(5))
title(main = "col = rainbow(5)")
barplot(vec01, names.arg = c(1,2,3,4,5), col = c("red", "green", "blue", "yellow", "black"))
title(main = "col = color vector")
```

```r
par(mfrow = c(1,3))
vec01 <- c(1,3,5,2,4)
barplot(vec01, names= c(1,2,3,4,5), col = "white", border = "red")
title(main = "default color, border = 'red'")
barplot(vec01, names= c(1,2,3,4,5), col = "white", border = rainbow(5))
title(main = "border = rainbow(5)")
barplot(vec01, names = c(1,2,3,4,5), col = "white", border = c("red", "green", "blue", "yellow", "black"))
```

```r
par(mfrow = c(1,1))
scores <- c(60, 80, 70, 100, 90, 50)
colors <- c()
for (i in 1:length(scores)) {
  if (scores[i] >= 90) {
    colors <- c(colors, "red")
  } else if (scores[i] >= 80) {
    colors <- c(colors, "blue")
  } else if (scores[i] >= 70) {
    colors <- c(colors, "green")
  } else if (scores[i] >= 60) {
    colors <- c(colors, "yellow")
  } else {
    colors <- c(colors, "black")
  }
}

barplot(scores, main = "scores", names.arg = c("이한림", "이이림", "이삼림", "이사림",
  "이오림", "이육림"), col = colors)
```
