```r
car <- read.csv('/content/sample_data/도로교통공단_시도_시군구별_월별_교통사고(2018).csv',fileEncoding = 'cp949')
car
```

```r
dim(car)
```

```r
summary(car)
```

```r
car %>% group_by(시도) %>% summarise(mean1= mean(발생건수))
```

```r
df1 <- car %>% group_by(시도) %>% summarise_each(funs(sum), 발생건수, 사망자수, 부상자수)
df1
```

```r
name01 <- df1$시도
name01
```

```r
df2 <- df1 %>% select(발생건수, 사망자수, 부상자수)
df2
```

```r
m1 <- as.matrix(df2)
m1
```

```r
barplot(m1, beside = T, col = rainbow(17))
```

```r
getwd()
```

```r
df3 <- car %>% filter(시군구 == "종로구")
df3
```

```r
barplot((df3$발생건수 ~ df3$월), col = rainbow(12), main = "종로구 월별 발생건수", xlab = "월", ylab = "발생건수")
```

```r
plot(df3$발생건수, type = "o", col = "red", bg = "yellow", axes = FALSE, ann = FALSE)
axis(1, at= 1:12, label = df3$월)
axis(2, ylim = c(df3$발생건수))
title(main = "종로구 발생건수", xlab = "월", ylab = "발생건수")
```

```r
v1 <- as.matrix(df2)[,1]
v1
```

```r
v2 <- as.matrix(df2)[,2]
v2
```

```r
v3 <- as.matrix(df2)[,3]
v3
```

```r
options(repr.plot.width=15, repr.plot.height=10)

plot(v1, type = "l", col = "red", lwd = 2, axes = F, ann = F)

lines(v2, type = "l", col = "blue", lwd= 2)
lines(v3, type = "l", col = "green", lwd = 2)
axis(1, at = 1:17, name01, las = 2)
axis(2, ylim = c(0:100000))
abline(h = seq(0, 100000, 10000), lty = 2, col = "red")
legend("topright", 55000, name01, cex = 0.6, lty = 1:17)
```

```r
pie(v1, col = rainbow(10), labels = name01, main = "발생건수")
```

```r
pct <- round(v1/sum(v1)*100, 1)
pct
```

```r
lab <- paste(name01, "\n", pct, "%")
lab
```

```r
length(name01)
```

```r
pie(v1, col = rainbow(length(name01)), main = "발생건수", cex = 0.8, labels = lab)
```
