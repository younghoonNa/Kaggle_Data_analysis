```r
print(x)
print(y)

par(mfrow = c(2,4))
#type
plot(x,y, type='l', xlab = "X lab" , ylab = "Y lab", col = 'blue', main = 'Graph') 
plot(x,y, type='p', xlab = 'X lab' , ylab = "y lab", col = 'blue', main = 'Graph2') #default - pointer
plot(x,y, type='b', xlab = 'X lab' , ylab = "y lab", col = 'blue', main = 'Graph3') #b 
plot(x,y, type='o', xlab = 'X lab' , ylab = "y lab", col = 'blue', main = 'Graph4') #o 
plot(x,y, type='h', xlab = 'X lab' , ylab = "y lab", col = 'blue', main = 'Graph5') #h
plot(x,y, type='s', xlab = 'X lab' , ylab = "y lab", col = 'blue', main = 'Graph6') #s
plot(x,y, type='S', xlab = 'X lab' , ylab = "y lab", col = 'blue', main = 'Graph7') #S
plot(x,y, type='n', xlab = 'X lab' , ylab = "y lab", col = 'blue', main = 'Graph8') #h
```

```r
print(x)
print(y)

par(mfrow = c(2,4))
#lty :  . (0 ~ 6) 
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', lty = 0, main = 'Graph_line0')
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', lty = 1, main = 'Graph_line1')
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', lty = 2, main = 'Graph_line2')
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', lty = 3, main = 'Graph_line3')
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', lty = 4, main = 'Graph_line4')
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', lty = 5, main = 'Graph_line5')
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', lty = 6, main = 'Graph_line6')
```

```r
print(x)
print(y)

par(mfrow = c(2,4))
# (lwd)
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', main = 'Graph_line0', lwd=0)
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', main = 'Graph_line1', lwd=2)
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', main = 'Graph_line2', lwd=4)
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', main = 'Graph_line3', lwd=6)
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', main = 'Graph_line4', lwd=8)
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', main = 'Graph_line5', lwd=10)
plot(x,y,type='l', xlab = "X lab", ylab = "y lab", col = 'red', main = 'Graph_line6', lwd=20)
```
