```r
par(mfrow = c(1,1))
scores <- c(60, 80, 70, 100, 90, 50)
colors <- c()
colors

for (i in 1:length(scores)) {
  if (scores[i] >= 90) {
    colors <- c(colors, "red")
  } else if (scores[i] >= 80 && scores[i] <= 90) {
    colors <- c(colors, "blue") 
  } else if (scores[i] >= 70 && scores[i] <= 80) {
    colors <- c(colors, "green")
  } else if (scores[i] >= 60 && scores[i] <= 70) {
    colors <- c(colors, 'yellow')
  } else {
    colors <- c(colors, "black")
  }
}

colors
barplot(scores, main = "Scores", names.arg = c("이한림", "이이림", "이삼림", "이사림", 
        "이오림", "이육림"), col = colors)
```

```r
scores <- c(80, 60, 70, 100, 90 ,50)
colors <- c()

for (i in 1:length(scores)) {
  ifelse(scores[i] >= 90 , colors <- c(colors, "red"),
    ifelse(scores[i] >= 80, colors <- c(colors, "blue"),
      ifelse(scores[i] >= 70, colors <- c(colors, "green"),
        ifelse( scores[i] >= 60, colors <- c(colors, "blue"),
          colors <- c(colors, "black")))))
}

colors

barplot(scores, main = "Scores", names.arg = c("이한림", "이이림", "이삼림",
      "이사림", "이오림", "이육림"), col = colors)
```

```r
vec01 <- c(1,3,5,2,4)
cols <- rainbow(length(vec01))
pct <- round(vec01 / sum(vec01) * 100, 1)
lab1 <- c("이한림", "이이림", "이삼림", "이사림", "이오림")
lab2 <- paste(lab1, "\n", pct, "%")
pie(vec01, radius = 1, col = cols, labels = lab2, main = "Labels and Percentages(-1, 1)")
legend(-1, 1, lab1, cex = 0.7, fill = cols)
```

```r
pie(vec01, radius = 1, col = cols, labels = lab2, 
    main = "Labels and Percentages (-1.6, 1)")
legend(-0.6, 1, lab1, cex = 0.7, fill = cols)
```

```r
df <- tat %>% filter(시도 == "서울") %>% group_by(시군구) %>% summarise(
  건수합 = sum(발생건수))
df
```

```r
danger_list = c()

for (i in 1:length(df$건수합)) {
  if (df$"건수합"[i] >= 3000) {
    danger_list <- c(danger_list, "red")
  } else if (df$"건수합"[i] >= 2500 && df$"건수합"[i] <= 3000) {
    danger_list <- c(danger_list, "blue")
  } else if (df$"건수합"[i] >= 2000 && df$"건수합"[i] <= 2500) {
    danger_list <- c(danger_list, "green")
  } else if (df$"건수합"[i] >= 1500 && df$"건수합"[i] <= 2000) {
    danger_list <- c(danger_list, "cyan")
  } else if (df$"건수합"[i] >= 1000 && df$"건수합"[i] <= 1500) {
    danger_list <- c(danger_list, "yellow") 
  } else {
    danger_list <- c(danger_list, "purple")
  }
}

barplot(df$`건수합`, names.arg = df$시군구, las = 2, ylim = c(0,3500),
  main = "2018년도 서울특별시의 교통사고 발생건수 지역별 합.", 
  col = danger_list)
abline(a=100, b=40, col = "red", lwd = 3)
grid(1,7, lwd=1, col = "black")
```
