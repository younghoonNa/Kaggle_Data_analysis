```r
plot(0:20, 0:20, type= "n", xlim = c(0,20), ylim = c(0,20), xlab = "x label", ylab = "y label", main = "Lines")

lines(c(1,15), c(20,20), lty = 0); text(17, 20, "0:blank")
lines(c(1,15), c(18,18), lty = 1); text(17, 18, "1:solid")
lines(c(1,15), c(16,16), lty = 2); text(17, 16, "2:dashed")
lines(c(1,15), c(14,14), lty = 3); text(17, 14, "3:dotted")
lines(c(1,15), c(12,12), lty = 4); text(17, 12, "4:dotdash")
lines(c(1,15), c(10,10), lty = 5); text(17, 10, "5:longdash")
lines(c(1,15), c(8,8), lty = 6); text(17, 8, "6:twodash")

# lines(c(1,15), c(6,6), lty = "BW"); text(17,6, "BW")
lines(c(1,15), c(4,4), lty = "2F"); text(17, 4, "2F")
lines(c(1,15), c(2,2), lty = "F2F9"); text(17,2, "F2F9")
```

```r
plot(1:20, 1:20, type ="n", axes = T, xlab = "x lab", ylab = "y lab", main = "arrows")
arrows(1,18, 14,18, angle = 30, length = 0.25, code = 2)
arrows(1,16, 14,16)
text(20,16, adj = 1, "default")

arrows(1,14, 14,14, length = 0.5)
text(20,14, adj = 0.7, "length = 0.5")

arrows(1,12, 14,12, length = 0.1)
text(20, 12, adj = 1, "length = 0.1")

arrows(1,10, 14,10, angle = 60)
text(20, 10, adj = 1, "angle = 60")

arrows(1,8, 14,8, angle = 90)
text(20, 8, adj = 1, "angle = 90")

arrows(1,6, 14,6, angle = 120)
text(20, 6, adj = 1, "angle = 1")

arrows(1,4, 14,4, code = 1)
text(20,4, adj = 1, "code = 1")

arrows(1,2, 14,2, code = 3)
text(20, 2, adj = 1, "code = 3")
```

```r
par(mfrow = c(2,1))
x <- 1:10
y <- c(3,2,7,4,5,9,2,4,1,8)
s=seq(length(x)-1)
plot(x,y, main = "1 segment")
segments(x[s], y[s], x[s+1], y[s+1], lty = 1)

plot(x,y, main = "2 segment")
segments(x[s], y[s], x[s+2], y[s+2], lty = 1:2)
```

```r
par(mfrow = c(2,1))
par(mar = c(2,2,2,2), oma = c(2,2,2,2))

x <- 1:10
y <- c(3,2,7,4,5,9,2,4,1,8)

plot(x,y,main = "1 segments")
s = seq(length(x))
segments(x[s], y[s], x[s+1], y[s+1], lty = 1)

box(which = 'outer', lty = 1, lwd = 1.2, col = "red")
box(which = 'inner', lty = 2, lwd = 1.4, col = "black")
box(which = 'figure', lty = 3, lwd = 1.6, col = "blue")
box(which = 'plot', lty = 4, lwd = 1.8, col = "green")
```
