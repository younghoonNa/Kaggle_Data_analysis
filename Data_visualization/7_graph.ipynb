```r
par(mfrow = c(1,1))
scores <- c(60, 80, 70, 100, 90, 50)
colors <- c()
colors

for (i in 1:length(scores)) {
  if (scores[i] >= 90) {
    colors <- c(colors, "red")
  } else if (scores[i] >= 80 && scores[i] <= 90) {
    colors <- c(colors, "blue") 
  } else if (scores[i] >= 70 && scores[i] <= 80) {
    colors <- c(colors, "green")
  } else if (scores[i] >= 60 && scores[i] <= 70) {
    colors <- c(colors, 'yellow')
  } else {
    colors <- c(colors, "black")
  }
}

colors
barplot(scores, main = "Scores", names.arg = c("이한림", "이이림", "이삼림", "이사림", 
        "이오림", "이육림"), col = colors)
```

```r
scores <- c(80, 60, 70, 100, 90 ,50)
colors <- c()

for (i in 1:length(scores)) {
  ifelse(scores[i] >= 90 , colors <- c(colors, "red"),
    ifelse(scores[i] >= 80, colors <- c(colors, "blue"),
      ifelse(scores[i] >= 70, colors <- c(colors, "green"),
        ifelse( scores[i] >= 60, colors <- c(colors, "blue"),
          colors <- c(colors, "black")))))
}

colors

barplot(scores, main = "Scores", names.arg = c("이한림", "이이림", "이삼림",
      "이사림", "이오림", "이육림"), col = colors)
```

```r
df <- tat %>% filter(시도 == "서울") %>% group_by(시군구) %>% summarise(mean = round(mean(발생건수)))
df
```

```r
danger_list = c()

for (i in 1:length(df$건수합)) {
  if (df$"건수합"[i] >= 3000) {
    danger_list <- c(danger_list, "red")
  } else if (df$"건수합"[i] >= 2500 && df$"건수합"[i] <= 3000) {
    danger_list <- c(danger_list, "blue")
  } else if (df$"건수합"[i] >= 2000 && df$"건수합"[i] <= 2500) {
    danger_list <- c(danger_list, "green")
  } else if (df$"건수합"[i] >= 1500 && df$"건수합"[i] <= 2000) {
    danger_list <- c(danger_list, "cyan")
  } else if (df$"건수합"[i] >= 1000 && df$"건수합"[i] <= 1500) {
    danger_list <- c(danger_list, "yellow") 
  } else {
    danger_list <- c(danger_list, "purple")
  }
}

barplot(df$`건수합`, names.arg = df$시군구, las = 2, ylim = c(0,3500),
  main = "2018년도 서울특별시의 교통사고 발생건수 지역별 합.", 
  col = danger_list)
grid(1,7, lwd=1, col = "black")

arrows(10,3400, 2,3400, col = "red", lwd = 3)
text(12, 3400, "가장 많은 수치를 기록한 시는 강남구", adj = 0.1)
```

```r
df_box <- tat %>% filter(시군구 == "강남구") %>% select(발생건수, 사망자수, 부상자수, 중상, 경상, 부상신고)
df_box
```

```r
boxplot(df_box, col = 1:6, width = 1:6, border = "red")
title(main = paste("2018년도 서울시 강남구의 주요 교통사고 통계", "\n", "출처[도로교통공단]"))
arrows(4,200, 1.1,249, lwd = 2, col = "red", angle = 25, length = 0.2, lty = 5)
text(4,240, paste("2,9월", "\n", "이상치"), col = "red", cex = 1.5, font = 2, adj = 0.7)

legend("bottomleft", c("발생건수", "사망자수", "부상자수", "중상", "경상", "부상신고"), fill = 1:6, density = 90)
```

```r
box_width = c()
box_color = c()

for (i in 1:6) {
  sum_of_box= sum(df_box[i])
  if (sum_of_box > 4000) {
    box_width <- c(box_width , 6)
    box_color = c(box_color, "red")
  } else if (sum_of_box > 3500 && sum_of_box < 4000) {
      box_width <- c(box_width , 5)
      box_color = c(box_color, "blue")
  } else if (sum_of_box > 3000 && sum_of_box < 3500) {
      box_width <- c(box_width , 4)
      box_color = c(box_color, "green")
  } else if (sum_of_box > 800 && sum_of_box < 3000) {
      box_width <- c(box_width , 3)
      box_color = c(box_color, "cyan")
  } else if (sum_of_box > 200 && sum_of_box < 800) {
      box_width <- c(box_width , 2)
      box_color = c(box_color, "yellow")
  } else {
      box_width <- c(box_width , 1)
      box_color = c(box_color, "black")  
  }
}

boxplot(df_box, col = box_color, width = box_width, border = "red", notch = TRUE)
title(main = paste("2018년도 서울시 강남구의 주요 교통사고 통계", "\n", "출처[도로교통공단]"))
arrows(4,200, 1.1,249, lwd = 2, col = "red", angle = 25, length = 0.2, lty = 5)
text(4,240, paste("2,9월", "\n", "이상치"), col = "red", cex = 1.5, font = 2, adj = 0.7)

arrows(5 , 420, 3.5, 420, lwd = 3, col = "green", angle = 25, length = 0.2)
text(5.2,420, paste("수치가 클수록", "\n","넓이가 큼."), col = "green", cex = 1.5, font = 2, adj = 0.1)

legend("bottomleft", c("발생건수", "사망자수", "부상자수", "중상", "경상", "부상신고"), fill = box_color, density = 90)
```

```r
df_pie <- tat %>% filter(시도 == "울산")  %>% group_by(시군구) %>% summarise(mean = round(mean(발생건수)))
df_pie
```

```r
pie_col <- c()

for (i in 1:length(df_pie$시군구)) {
  if (df_pie$mean[i] > 100) {
    pie_col <- c(pie_col , "red")
  } else if (df_pie$mean[i] >= 65 && df_pie$mean[i] <= 100) {
    pie_col <- c(pie_col , "blue") 

  } else if (df_pie$mean[i] >= 55 && df_pie$mean[i] <= 65) {
    pie_col <- c(pie_col , "green") 

  } else if (df_pie$mean[i] >= 45 && df_pie$mean[i] <= 65) {
    pie_col <- c(pie_col , "yellow") 

  } else {
    pie_col <- c(pie_col , "cyan") 

  }
}

pie(df_pie$mean, clockwise = T, init.angle = 0, col = pie_col, labels = 
      paste(df_pie$`시군구`, "\n", round(df_pie$mean/sum(df_pie$mean) * 100,2),"%"),
      font= 2, cex = 2, density = 60)
points(0,0, cex=5, col = "white", lwd = 200)
title("2018년도 울산시의 시군구별 교통사고 평균")
arrows(0,0,0.4,-0.4, col = "purple", lwd = 3)
text(0,0, pos = 3, "가장 많은 사고 발생", cex = 1.2, font = 4)
```
