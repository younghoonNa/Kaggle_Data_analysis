```r
par(mfrow=c(1,1), mar = c(2,2,2,2)) #margin -> 아래 왼 위 오 순으로 
plot(1:20, 1:20, type = 'n', main = "Font")
for (i in 1:5) {
  par(font = i)
  text(10, 20-i*3, paste("computer", i), cex=2) #paste는 붙이다 라는 것.
}
```

```r
par(mfrow=c(1,1), mar=c(2,2,2,2), bty = 'l') #bty = box type
plot(0:20, 0:20, type = 'n')
text(13,10, "computer 0 ", srt = 0, cex=1.5)
text(10,15, "computer 90 ", srt = 90, cex=1.5)
text(6,10, "computer 180", srt = 180, cex=1.5)
text(10,5, "computer 270", srt = 270, cex=1.5)
```

```r
x <- c(1,2,3,4,5)
y <- c(1,3,5,2,4)
par(mfrow = c(1,2))
par("mar")
```

```r
par(mar = c(10,5,2,2))
plot(x,y, type='l', main = "mar = c(10,5,2,2)") #아래쪽 마진 10
```

```r
par(mar = c(2,2,10,5))
plot(x,y, type ='l', main = "mar = c(2,2,10,5)")
```

```r
vec01 <- c(1:5)
vec02 <- c(1,3,2,5,4)
par(oma = c(3,0,0,0))
plot(vec01, vec02, type = 'l', xlab = 'x label', main = 'oma = c(2,1,0,0)')
```

```r
par(oma=c(0,3,0,0), new = T)
plot(vec01, vec02, type = 'l', xlab = "x label", main = "oma = c(0,2,0,0)")
```

```r
par(mfrow = c(1,2))
par(oma = c(0,3,0, 0))
plot(vec01, vec02, type = 'l', xlab = "x label", main = "oma = c(0,3,0,0)")

par(oma = c(3,0,0,0), new = T)
plot(vec01, vec02, type = 'l', xlab = "x label", main = "oma = c(3,0,0,0)")
```

```r
par(mfrow = c(1,2))
vec01 <- c(1,3,2,5,4)
plot(vec01, type = 'l', xlab = "x label", main = "mgp = Default")
par(mgp=c(4,0,0)) #mgp = c(제목 위치, 지표값 위치, 지표선 위치)
plot(vec01, type = 'l', xlab = "x label", main = "mgp = c(4,0,0)")
```

```r
par(mfrow = c(1,2))
vec01 <- c(1,3,2,5,4)
plot(vec01, type = 'l', xlab = "x label", main = "mgp = Default")
par(mgp=c(0,4,0)) #mgp = c(제목 위치, 지표값 위치, 지표선 위치)
plot(vec01, type = 'l', xlab = "x label", main = "mgp = c(0,4,0)")
```

```r
par(mfrow = c(1,2))
vec01 <- c(1,3,2,5,4)
plot(vec01, type = 'l', xlab = "x label", main = "mgp = Default")
par(mgp=c(0,0,3)) #mgp = c(제목 위치, 지표값 위치, 지표선 위치)
plot(vec01, type = 'l', xlab = "x label", main = "mgp = c(0,0,3)")
```

```r
par(mfrow= c(1,2))
vec01 <- c(1,3,2,5,4)

par(mgp = c(3,2,1))
plot(vec01 , type = 'l', xlab = 'x label', main = "mgp = c(3,2,1)")
par(mgp = c(1,2,3))
plot(vec01, type ='l', xlab = 'x label', main = "mgp = c(1,2,3)")
```

```r
x <- c(1:5)
y <- c(1,3,5,2,4)

par(mfrow = c(2,3), bty = 'l')
plot(x, y, main = "bty = \"l\"", type = 'l')
plot(x, y, main = "bty = \"c\"", type = 'l', bty = 'c')
plot(x, y, main = "bty = \"n\"", type = 'l', bty = 'n')
plot(x, y, main = "bty = \"o\"", type = 'l', bty = 'o')
plot(x, y, main = "bty = \"7\"", type = 'l', bty = '7')
plot(x, y, main = "bty = \"u\"", type = 'l', bty = 'u')
```

```r
theta <- seq(-pi, pi, length = 30) #점은 29개 why -> 시작과 끝이 같이 때문에.
x <- cos(theta) * 10
y <- sin(theta) * 10

par(mfrow=c(1,2), pty = 's', bty = 'o')
plot(x, y, type = 'b', main = "pty = s")
par(pty = 'm')
plot(x,y, type = 'b', main = 'pty = m')
```

```r
par(pty = 's')

theta <- seq(-pi, pi, length = 361)
x <- 10 + 5 * cos(theta)
y <- 10 + 5 * sin(theta)

plot(x,y, type = 'l', main = "Circle Radius = 5, start from - (10,10)")
```

```r
par(pty = 'm')
theta <- (0:360) * pi / 180
x <- 10 + 5 * cos(theta)
y <- 10 + 5 * sin(theta)
plot(x,y,type = 'b', main = "Ellipse")
```
