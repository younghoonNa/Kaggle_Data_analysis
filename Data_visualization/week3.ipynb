```r
ctNames <- c("폴란드","영국","인도","프랑스","소련","유고슬라비아","미국","독일","일본","이탈리아")
milDeaths <- c(240,383.7,87,210,8668,300,407.3,4440,2100,319.2)
milWounds <- c(766.606,376.239,390,64.354,14685.593,425,671.801,7300,326,225)
civDeaths <- c(5620,67.2,2100,390,4500,581,12.1,1500,550,153.2)
totDeaths <- c(5900,450.9,2200,600,20000,1027,419.4,6900,2500,492.4)
df_01 <- data.frame(국가=ctNames,군사사망자=milDeaths,군사부상자=milWounds,민간사망자=civDeaths,종합사망자=totDeaths)
```

```r
options(repr.plot.width=10, repr.plot.height=8)

plot(df_01$군사사망자, axes = FALSE, ann = FALSE, type = 'l', lwd = 2, col = 'red', ylim = c(0, 20000))
axis(1, at=1:10, label = ctNames)
axis(2)
title(main = "제 2차 세계대전 국가별 부상 및 사망자 수", xlab = "국가들", ylab = "범위")
lines(df_01$군사부상자, col = 'blue', lwd = 2,)
lines(df_01$민간사망자, col = 'green', lwd = 2,)
lines(df_01$종합사망자, col = 'cyan', lwd = 2,)
legend( 1, 20000, c("군사사망자", "군사부상자", "민간사망자", "종합사망자"),  col = c("red", 'blue', 'green', 'cyan'), lty=1, )
```

```r
options(repr.plot.width=10, repr.plot.height=8)

plot(df_01$군사사망자,type = 's', lwd = 2, col = 'red', ylim = c(0, 20000), main = "제 2차 세계대전 국가별 부상 및 사망자 수를 다양한선 그래프로 표현하기.",
         xlab = "국가들", ylab = "범위")
axis(1, at=1:10)
par(new = TRUE)
plot(df_01$군사부상자, col = 'blue', lwd = 2, type = 'b', ann = FALSE, axes = FALSE, ylim = c(0,20000))
par(new = TRUE)
plot(df_01$민간사망자, col = 'green', lwd = 2, type = 'o', ann = FALSE, axes = FALSE, ylim = c(0,20000))
par(new = TRUE)
plot(df_01$종합사망자, col = 'cyan', lwd = 2, type = 'c', ann = FALSE, axes = FALSE, ylim = c(0,20000))
legend( 1, 20000, c("군사사망자", "군사부상자", "민간사망자", "종합사망자"),  col = c("red", 'blue', 'green', 'cyan'), lty=1, bg = 'purple')
```

```r
# [D3-13] 야구 역대 최고기록 투수
stName <- c("최동원", "주형광", "장명부", "에르난데스", "선동열", "선동열", "류현진", "최동원", "스트레일리", "류현진")
vec01 <- c(223, 221, 220, 215, 214, 210, 210, 208, 205, 204) #탈삼진
vec02 <- c(30, 27, 25, 25, 24, 24, 23, 22, 22, 22) #승률
vec03 <- c(52, 51, 47, 47, 44, 42, 42, 41, 40, 40) #세이브포인트
vec04 <- c(47, 47, 46, 42, 40, 40, 39, 39, 38, 38) #세이브

df_02 <- data.frame(투수=stName, 탈삼진=vec01, 승률=vec02, 세이브포인트=vec03, 세이브=vec04)
```

```r
options(repr.plot.width=10, repr.plot.height=8)

plot(df_02$탈삼진, axes = FALSE, ann = FALSE, type = 'l', lwd = 2, col = 1, ylim = c(0, 300))
axis(1, at=1:10, label = stName)
axis(2)
title(main = "야구 역대 최고기록 투수", xlab = "투수", ylab = "기록")
lines(df_02$승률, col = 2, lwd = 2,)
lines(df_02$세이브포인트, col = 3, lwd = 2,)
lines(df_02$세이브, col = 4, lwd = 2,)
legend( 1, 200, c("탈삼진", "승률", "세이브포인트", "세이브"),  col = c(1:4), lty=1)
```

```r

st1Name <- c("미국", "독일", "영국", "프랑스", "이탈리아", "캐나다", "한국", "호주", "스페인", "멕시코")
education <- c(91, 87, 81, 78, 61, 91, 88, 81, 59, 38) #교육수준
vote <- c(65, 76, 69, 75, 73, 68, 77, 91, 70, 63) # 투표율
employment <- c(70, 75, 75, 65, 58, 73, 67, 73, 62, 61) #고용률
house <- c(19, 20, 26, 21, 23, 22, 15, 20, 21, 20) # 주택비율

df_03 <- data.frame(국가 = st1Name, 교육수준 = education, 투표율 = vote, 고용률 = employment, 주택비율 = house)
```
