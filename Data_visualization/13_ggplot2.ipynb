## ggplot2 -> 증분 방식을 통해 R 기본 그래픽스보다 인터랙티브 하며 훨씬 쉽고 다양하게 표현 가능.!

```r
if (!require(ggplot2)) {
  install.pacages("ggplot2")
  require(ggplot2)
}
```

```r
head(mtcars)
```

```r
# mpg : 1개론당 갈 수 있는 마일로 자동차 연비
# cyl : 실린더의 수, 4기통, 6기통.. ect
# disp : 배기량 : 1500cc, 1800cc, 2000cc
# hp : 마력
# drat : Rear axle ratio
# wt : 자동차의 무게
# qsec (1/4 mile time)
# vs(Engine ( 0  = V-shaped, 1 = staright))
# am(Transmission ( 0 = automatic, 1 = manual)) : 변속기가 자동인지 아니면 수동인지
# gear(Number of forward gears) : 기어 -> 기어 3단 4단 5단
# carb(Number of carburetors) : 카뷰레터는 공기와 휘발유를 섞은 액체를 엔진의 흡기 매니폴드로 전달시켜 엔진의 4행정을 도와주는 물건.
# cyl2 : 실린더 방식
```

```r
ggplot() + geom_point(iris, mapping = aes(x = Sepal.Length, y = Sepal.Width, colour = Species)) # Sepal : 꽃받침
# 세토사는 꽃받침의 길이가 짧고 폭이 넓음
# 버지칼라는 길이와 폭이 모두 짧다.
# 버지니카는 꽃받침의 길이와 폭이 모두 큼
```

```r
ggplot() + geom_point(iris, mapping = aes(x = Sepal.Length, y = Sepal.Width, colour = Species)) + geom_smooth (iris, mapping = aes(x = Sepal.Length, y=  Sepal.Width))
# geom_smooth()를 통해 전체적인 분포 확인 가능.
```

```r
g1 <- ggplot(data = diamonds, aes(carat, y = price))
```

```r
g1 + geom_point() # 무게가 무거울수록 가격이 올라감.
```

```r
g1 + geom_point(alpha = 0.1, color= 'blue') # alpha 는 투명도 설정
```

```r
diamonds_plot <- ggplot(data = diamonds, aes(x= carat, y = price))
diamonds_plot + geom_point(alpha = 0.1, aes(color = cut)) # 자른 단면과 가격 비교
```

```r
ggplot(diamonds , aes( x= cut , fill = cut)) + geom_bar()
```

```r
ggplot(diamonds, aes ( x = cut, fill = cut)) + geom_bar()  +scale_fill_brewer()
# scale_fill_brewer을 통해 색상 변경
```

```r
ggplot(diamonds, aes ( x = cut, fill = cut)) + geom_bar()  +scale_fill_brewer(palette = "Accent")
# scale_fill_brewer을 통해 색상 변경
```

```r
ggplot(diamonds, aes(x = cut, fill = cut)) + geom_bar() + coord_flip()
```

```r
ggplot(diamonds, aes(x = cut, fill = cut)) + geom_bar() + coord_polar() #극좌표로 변환.
```
