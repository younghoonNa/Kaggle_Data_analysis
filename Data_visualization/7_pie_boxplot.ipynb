```r
par(mfrow = c(1,2), oma = c(0.5, 0.5, 0.1, 0.1))
vec01 <- c(1,3,5,2,4)
pie(vec01, main = "Radius = 0.8 (default)")
pie(vec01, main = "Radius = 1", radius = 1)
```

```r
vec01 <- c(1,3,5,2,4)
par(mfrow = c(1,2))
pie(vec01, radius = 1, main = "Angle = default")
pie(vec01, radius = 1, init.angle = 90, main = "Angle = 90")
```

```r
vec01 <- c(1,3,5,2,4)
par(mfrow = c(1,2))
pie(vec01, radius = 1, main = "Clockwise = default")
pie(vec01, radius = 1, clockwise = TRUE, main = "Clockwise = True")
```

```r
vec01 <- c(1,3,5,2,4)
pie(vec01, radius = 1, col = rainbow(length(vec01)), labels = c("first", "second", "thrid", "fourth", "fifth"),
    main = "Color and Labels")
```

```r
vec01 <- c(1,3,5,2,4)
pie(vec01, density = 30, angle = 70, radius = 1, col = rainbow(length(vec01)),
    labels = c("first", "second", "third", "fourth", "fifth",
    main = "Color and Labels"))
```

```r
vec01 <- c(1,3,5,2,4)
cols <- rainbow(length(vec01))
pct <- round(vec01/sum(vec01) * 100, 1)
lab1 <- c("first", "second", "thrid", "fourth", "fifth")
lab2 <- paste(lab1, "\n", pct, "%")
pie(vec01, radius = 1, col = cols, labels = lab2, main = "Labels and Percentages")
legend(-1, 1, lab1, cex =1, fill = cols)
```

```r
install.packages("plotrix")
library(plotrix)

vec01 <- c(1,3,5,2,4)
cols <- c("red", "blue", "green", "yellow", "black")
pct <- round(vec01/sum(vec01)*100 ,1)
lab1 <- c("first", "second", "thrid", "fourth", "fifth")
lab2 <- paste(lab1, "\n", pct, "%")

pie3D(vec01, radius = 0.6, col = cols, labels = lab2, cex=  0.6, explode = 0.05, 
  main = "Labels and Percentages")
legend(-0.2, 1 , lab1, cex = 0.6, fill =cols)
```

```r
par(mfrow = c(1,2))
vec01 <- c(1,3,5,2,4)
boxplot(vec01)
# vec01
vec02 <- c(21,23,22,25 ,24)
vec03 <- c(11,13,12,15,14)
boxplot(vec01, vec02, vec03)
```

```r
par(mfrow = c(1,2))
vec01 <- c(1,3,2,5,4,6,7,8,9)
vec02 <- c(15,16,17,18,19,21,23,22,25,24)
vec03 <- c(9,10,11,13,12,15,14,16,17,18)

boxplot(vec01, vec02, vec03, col = c('red', 'yellow', 'green'))
boxplot(vec01, vec02, vec03, col = c("red", "yellow", 'green'), border = "blue",
    names = c('first' , ' second' , 'thrid'), horizontal = T)
```

```r
vec01 <- c(1,3,2,5,4,6,7,8,9,20)
par(mfrow = c(1,2))
boxplot(vec01, main = "range = default(1.5)")
boxplot(vec01, range = 0, main = "range = 0")
```

```r
par(mfrow = c(1,2))
vec01 <- c(1,3,2,5,4,6,7,8,9,20)
boxplot(vec01, main = "outline = default(TRUE)")
boxplot(vec01, outline = FALSE, main = "outline = False")
```

```r
par(mfrow = c(1,2))
vec01 <- c(1,3,2,5,4,6,7,8,9)
vec02 <- c(15,16,17,18,19,21,23,22,25,24)
vec03 <- c(9,10,11,13,12,15,14,16,17,18)

x <- list(vec01, vec02, vec03)

boxplot(x, width = 1:3, col = c("red", "yellow", "green"))
boxplot(vec01, vec02, vec03, width = 1:3, col = c("red", "yellow", "green"), 
  main = "Main", border = "blue", names = c("first", "second", "third"),las = 2)
```

```r
vec01 <- c(1,3,2,5,4,6,7,8,9,20)
boxplot(vec01, main = "notch = default(FALSE)")
boxplot(vec01, notch = TRUE, main = "notch = TRUE")
```
